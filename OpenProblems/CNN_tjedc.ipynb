```

def Read():
    X_train = []
    Path = '/content/drive/My Drive/question2/7632a1cf-95f9-4d1c-bc5b-9436699bdae2_train/ntrain/'
    Information = pd.read_csv('/content/drive/My Drive/question2/db5fcca9-f52b-42a9-87be-26f77b6f9d97_train.csv')
    Lines = Information['image_file']
    Y_train = Information['emotion'] 
    print(len(Lines))
    i=0
    for line in Lines:
        # print('Reading..........')
        path = Path + line+'.jpg'
        img = cv2.imread(path)
        if img is None:
            continue
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        scale_percent = 0.6
        width = int(gray.shape[1]*scale_percent)
        height= int(gray.shape[0]*scale_percent)
        dimension = (width, height)
        resized = cv2.resize(gray, dimension, cv2.INTER_AREA)
        # print(resized.shape)
        # resized = resized.flatten()
        X_train.append(resized)
        print(i)
        i=i+1
    X_train = np.array(X_train)
    Y_train = np.array(Y_train)
    return X_train, Y_train


def ReadTest():
    X_test = []
    Path = '/content/drive/My Drive/test/'
    Information = pd.read_csv('/content/c3b70bec-470d-4981-82ea-6e0693f2c8b0_test(1).csv')
    Lines = Information['image_file']
    print(Lines)
    i = 0
    for line in Lines:
        # print('Reading........')
        path = Path + line + '.jpg'
        print(path)
        img = cv2.imread(path)
        if img is None:
            continue
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        scale_percent = 0.6
        width = int(gray.shape[1]*scale_percent)
        height= int(gray.shape[0]*scale_percent)
        dimension = (width,height)
        resized = cv2.resize(gray, dimension, cv2.INTER_AREA)
        # resized = resized.flatten()
        X_test.append(resized)
        print(i)
        i=i+1
    X_test = np.array(X_test)
    return X_test
```

```
class CNN_Classifier:
  
  model_CNN = Sequential()

  def standarization(self,X):
    X = (X - X.mean())/X.std()
    return X

  def predict(self,Xtest):
    classes   = self.model_CNN.predict_classes(Xtest)
    print(classes)
    return classes

  def train(self,X,y):

    X = X/255
    print(X.shape)
    print(y.shape)
    X1_train, X1_test, Y1_train, Y1_test = train_test_split(X, y, test_size=.1,shuffle= True, random_state=42)
    X2_train = X1_train.reshape(X1_train.shape[0],X1_train.shape[1],X1_train.shape[2],1).astype('float32')
    X2_test = X1_test.reshape(X1_test.shape[0],X1_test.shape[1],X1_test.shape[2],1).astype('float32')
    print(X2_train,X2_test)
    Y2_train =to_categorical(Y1_train,5)
    Y2_test = to_categorical(Y1_test,5)   
    # X1_train = X
    # X2_train = X1_train.reshape(X1_train.shape[0],X1_train.shape[1],X1_train.shape[2],1).astype('float32')
    # Y1_train  = y
    # Y2_train =to_categorical(y,5)  
    self.model_CNN.add(Conv2D(128, kernel_size=3, activation='relu', input_shape=(X2_train.shape[1],X2_train.shape[2],1)))
    self.model_CNN.add(MaxPooling2D(pool_size=(2, 2)))
    self.model_CNN.add(Conv2D(128, kernel_size=3, activation='relu'))
    self.model_CNN.add(MaxPooling2D(pool_size=(2, 2)))
    self.model_CNN.add(Conv2D(128, kernel_size=3, activation='relu'))
    self.model_CNN.add(MaxPooling2D(pool_size=(2, 2)))
    self.model_CNN.add(Flatten())
    self.model_CNN.add(Dense(5, activation='softmax'))
    self.model_CNN.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    self.model_CNN.fit(X2_train, Y2_train, epochs=20,verbose=True)
    print(X1_test.shape)
    classes = self.predict(X2_test)
    acc = accuracy_score(Y1_test,classes)
    print(classification_report(Y1_test,classes))
    print(acc)
    print(f1_score(Y1_test,classes,average = 'macro'))
  def test(self,X):

    X = X/255
    X2 = X.reshape(X.shape[0],X.shape[1],X.shape[2],1).astype('float32')
    print(X2.shape)
    return self.predict(X2)
```
